```python
# --- Part 2: Data Structures ---
# --- Task 1: Explore the Menu ---

# Provided Data
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

# 1. Print formatted menu grouped by category
categories = ["Starters", "Mains", "Desserts"]

for cat in categories:
    print(f"\n===== {cat} =====")
    for item, details in menu.items():
        if details["category"] == cat:
            availability = "Available" if details["available"] else "Unavailable"
            # Using f-string padding for alignment
            print(f"{item:<18} ₹{details['price']:>6.2f}   [{availability}]")

# 2. Menu Statistics using Dictionary Methods
print("\n" + "="*30)
print("      MENU STATISTICS")
print("="*30)

# Total number of items
total_items = len(menu)

# Total number of available items
available_items_count = sum(1 for item in menu.values() if item["available"])

# Most expensive item
# Using the 'key' argument in max() to compare prices within the nested dictionaries
most_expensive_name = max(menu, key=lambda x: menu[x]["price"])
most_expensive_price = menu[most_expensive_name]["price"]

# Items priced under ₹150
affordable_items = [f"{name} (₹{data['price']})" for name, data in menu.items() if data["price"] < 150]

print(f"Total Items on Menu     : {total_items}")
print(f"Total Available Items   : {available_items_count}")
print(f"Most Expensive Item     : {most_expensive_name} (₹{most_expensive_price})")
print(f"Items under ₹150        : {', '.join(affordable_items)}")
```

An if condition filters the items belonging to the current category.

2. Advanced Dictionary Methods & Aggregation
len(menu): Quickly gives the count of top-level keys.

Generator Expressions: sum(1 for item in menu.values() if item["available"]) is a memory-efficient way to count items matching a specific internal criteria without creating a new list.

max() with key: Instead of writing a manual loop to find the highest price, we used max(menu, key=...). This tells Python to look at the price field inside the nested dictionary while returning the name of the item.

3. Formatting and Alignment

:>6.2f: Right-aligns the price, ensuring 2 decimal places for currency consistency.

```python
# --- Part 2: Data Structures ---
# --- Task 2: Cart Operations ---

# Re-using the menu from Task 1 (Ensure this is in the same notebook)
cart = []

def add_to_cart(item_name, qty):
    # 1. Check if item exists in menu
    if item_name not in menu:
        print(f"❌ Error: '{item_name}' is not on the menu.")
        return
    
    # 2. Check if item is available
    if not menu[item_name]["available"]:
        print(f"❌ Error: '{item_name}' is currently unavailable.")
        return
    
    # 3. Check if already in cart to update quantity
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += qty
            print(f"Updated: {item_name} quantity increased by {qty}.")
            return
            
    # 4. If not in cart, add new entry
    new_entry = {
        "item": item_name, 
        "quantity": qty, 
        "price": menu[item_name]["price"]
    }
    cart.append(new_entry)
    print(f"Added: {item_name} x{qty} to cart.")

def remove_from_cart(item_name):
    for i, entry in enumerate(cart):
        if entry["item"] == item_name:
            cart.pop(i)
            print(f"Removed: {item_name} from cart.")
            return
    print(f"⚠️ Notice: {item_name} was not found in your cart.")

# --- Simulation Sequence ---

print("--- Starting Simulation ---")
# 1. Add "Paneer Tikka" × 2
add_to_cart("Paneer Tikka", 2)

# 2. Add "Gulab Jamun" × 1
add_to_cart("Gulab Jamun", 1)

# 3. Add "Paneer Tikka" × 1 (Should update quantity to 3)
add_to_cart("Paneer Tikka", 1)

# 4. Try "Mystery Burger" (Non-existent)
add_to_cart("Mystery Burger", 1)

# 5. Try "Chicken Wings" (Unavailable)
add_to_cart("Chicken Wings", 1)

# 6. Remove "Gulab Jamun"
remove_from_cart("Gulab Jamun")

# --- Print Final Order Summary ---
print("\n" + "="*35)
print(f"{'Order Summary':^35}")
print("="*35)

subtotal = 0
for entry in cart:
    item_total = entry["quantity"] * entry["price"]
    subtotal += item_total
    print(f"{entry['item']:<18} x{entry['quantity']}    ₹{item_total:>7.2f}")

gst = subtotal * 0.05
total_payable = subtotal + gst

print("-" * 35)
print(f"{'Subtotal:':<22} ₹{subtotal:>8.2f}")
print(f"{'GST (5%):':<22} ₹{gst:>8.2f}")
print(f"{'Total Payable:':<22} ₹{total_payable:>8.2f}")
print("="*35)
```

Iterative Accumulation: The subtotal is calculated by looping through the cart and multiplying quantity * price for each entry.

```python
import copy

# --- Task 3: Inventory Tracker with Deep Copy ---

# Initial Inventory Data (as provided)
inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

# 1. Deep Copy for Backup
inventory_backup = copy.deepcopy(inventory)

# Demonstration of Independence
inventory["Paneer Tikka"]["stock"] = 999
print(f"Demo - Current Stock (Paneer): {inventory['Paneer Tikka']['stock']}")
print(f"Demo - Backup Stock (Paneer) : {inventory_backup['Paneer Tikka']['stock']}")

# Restore to original state
inventory = copy.deepcopy(inventory_backup)

# 2. Simulate Order Fulfilment (Using cart from Task 2)
# Current cart contains: Paneer Tikka x3
print("\n--- Processing Inventory Deductions ---")
for entry in cart:
    item = entry["item"]
    qty_ordered = entry["quantity"]
    
    if item in inventory:
        current_stock = inventory[item]["stock"]
        if current_stock >= qty_ordered:
            inventory[item]["stock"] -= qty_ordered
            print(f"Deducted {qty_ordered} {item}(s). Remaining: {inventory[item]['stock']}")
        else:
            print(f"⚠️ Warning: Insufficient stock for {item}. Only {current_stock} available.")
            inventory[item]["stock"] = 0

# 3. Reorder Alerts
print("\n--- Reorder Alerts ---")
for item, data in inventory.items():
    if data["stock"] <= data["reorder_level"]:
        print(f"⚠️ Reorder Alert: {item} — Only {data['stock']} unit(s) left (reorder level: {data['reorder_level']})")


# --- Task 4: Daily Sales Log Analysis ---

sales_log = {
    "2025-01-01": [{"order_id": 1, "items": ["Paneer Tikka", "Garlic Naan"], "total": 220.0}, {"order_id": 2, "items": ["Gulab Jamun", "Veg Soup"], "total": 210.0}, {"order_id": 3, "items": ["Butter Chicken", "Garlic Naan"], "total": 360.0}],
    "2025-01-02": [{"order_id": 4, "items": ["Dal Tadka", "Garlic Naan"], "total": 220.0}, {"order_id": 5, "items": ["Veg Biryani", "Gulab Jamun"], "total": 340.0}],
    "2025-01-03": [{"order_id": 6, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0}, {"order_id": 7, "items": ["Butter Chicken", "Veg Biryani"], "total": 570.0}, {"order_id": 8, "items": ["Garlic Naan", "Gulab Jamun"], "total": 130.0}],
    "2025-01-04": [{"order_id": 9, "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0}, {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"], "total": 270.0}],
}

# 1. Add New Day Data
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0}
]

# 2. Revenue per Day and Best Selling Day
daily_totals = {}
item_frequency = {}

print("\n--- Daily Revenue Report ---")
for date, orders in sales_log.items():
    day_revenue = sum(order["total"] for order in orders)
    daily_totals[date] = day_revenue
    print(f"{date}: ₹{day_revenue:>7.2f}")
    
    # Tracking item frequency for most ordered item
    for order in orders:
        for item in order["items"]:
            item_frequency[item] = item_frequency.get(item, 0) + 1

best_day = max(daily_totals, key=daily_totals.get)
most_ordered = max(item_frequency, key=item_frequency.get)

print(f"\nBest Selling Day : {best_day} (₹{daily_totals[best_day]})")
print(f"Most Ordered Item: {most_ordered} (Ordered {item_frequency[most_ordered]} times)")

# 3. Numbered List of All Orders
print("\n--- Complete Order History ---")
counter = 1
for date, orders in sales_log.items():
    for order in orders:
        items_str = ", ".join(order["items"])
        print(f"{counter}. [{date}] Order #{order['order_id']} — ₹{order['total']:.2f} — Items: {items_str}")
        counter += 1
```

Technical Explanation: Inventory & Sales Log (Part 2 - Tasks 3 & 4)
These tasks cover advanced data manipulation and the importance of memory management in Python.


Deep Copy Verification: > In Python, nested dictionaries are stored by reference. A simple assignment (backup = inventory) or a shallow copy (inventory.copy()) would link the two, meaning changing a stock value in one would change it in the other.

By using copy.deepcopy(), we ensure that the inventory_backup is stored in a separate memory location. As demonstrated in the output, modifying the stock in the inventory dictionary did not impact the inventory_backup, confirming a successful independent backup.

1. Deep Copy vs. Shallow Copy
In Python, nested dictionaries are stored as references. A standard "shallow copy" (like dict.copy()) only copies the top-level keys; the inner dictionaries would still point to the same memory location.

copy.deepcopy(): Creates a completely independent clone of the entire data structure. This ensures that changes to inventory do not corrupt inventory_backup.

2. Order Fulfilment Logic
We applied inventory depletion logic:

The script iterates through the cart and checks the inventory stock levels.

Constraint Handling: We ensured stock never drops below zero by using a conditional check, which is a standard safety protocol in ERP (Enterprise Resource Planning) systems.

3. Aggregating Nested Data
Task 4 involves "Flattening" and "Aggregating" data:

Revenue Calculation: We used a generator expression inside sum() to calculate total daily earnings.

Frequency Mapping: A dictionary (item_frequency) was used as a counter. The .get(item, 0) method is a clean way to handle new keys without triggering a KeyError.

Identifying Extremes: The max() function with a key argument (e.g., key=daily_totals.get) allows us to find the key (Date) associated with the highest value (Revenue).

4. Nested Iteration with enumerate
To generate a chronological master list of orders, we nested a loop for orders inside a loop for dates. By maintaining a global counter, we provided a flat, numbered view of a hierarchical dataset.
